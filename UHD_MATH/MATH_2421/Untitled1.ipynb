```julia
using Plots
```

```julia
a(n) = n^2/(2^n - 1)
```

```julia
points = [(n, a(n)) for n = 1:10]
```

```julia
plot(x -> x^2, label = "y = x^2", xlims = (0, 10))
plot!(x -> 2^x-1, label = "y = 2^x - 1")
```

```julia
function factorial(n)
    if n == 0 
        return 1
    elseif n == 1
        return 1
    else
        return n*factorial(n-1)
    end
end

c(n) = (-1)^n*(1/factorial(n))
```

```julia
points = [(n, c(n)) for n = 1:10]
scatter(points, label = "c_n")
scatter!([(n, -1/2^n) for n = 1:10], label = "b_n = -1/2^n")
scatter!([(n, 1/2^n) for n = 1:10], label = "a_n = 1/2^n")
```

```julia
for n = 1:10
    println("n = ", n)
    println("c_n = ", c(n))
    println("")
end
```
