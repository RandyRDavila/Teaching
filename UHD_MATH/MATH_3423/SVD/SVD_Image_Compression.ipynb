# SVD and Image Compression 
In this notebook we show how the SVD can be used to compress images. With this aim in mind, it should be noted that images can be represented as multidimensional arrays of numbers, i.e., matrices. 

```julia
# LinearAlgebra import the svd factorization function
# Images imports the functions for converting a standard image to a matrix
# TestImages is used to import a test image
using LinearAlgebra, Images, TestImages
```

```julia
# Load a sample test image 
img = testimage("cameraman")
```

```julia
# Convert the image to a matrix pf floating point numbers
A = float(channelview(img))
```

Next we use the svd function from the LinearAlgebra library to find the SVD factorization of A.

```julia
# Compute the SVD-factorization 
F = svd(A)
```

```julia
# Set the diagonal
 = Diagonal(F.S)
```
