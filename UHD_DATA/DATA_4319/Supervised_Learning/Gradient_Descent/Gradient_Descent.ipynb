# Gradient Descent 

Many machine learning algorithms are formulated in terms minimizing a given cost function. Let $C(w)$ denote some differentiable function of $w$. We wish to solve the following minimization problem. 

$$
\underset{w\in\mathbb{R}^n}\min C(w).
$$

One of the most popular techniques for such minimization is called *gradient descent*. This iterative method updates potential minimizers $w$ according the the following rule: At the $k$-th iteration of the algorithm, choose $w_{k+1}$ by the following equation

$$
w_{k+1} = w_k + \alpha_k \nabla C(w_k).
$$

Here, $\alpha_k$ is called the *step length* at iteration $k$ and $\nabla C(w_k)$ is the gradient of $C(w_k)$. Recall that for functions of a single variable, the gradient is simply the derivative. In this notebook we describe the gradient descent method by exploring some simple 2-dimensional plots. 

For this notebook we will need the following packages:
 * Plots [documentation](http://docs.juliaplots.org/latest/)

---

```julia
using Plots
theme(:dark)


# Define a simple function to explore minimization on.
C(w) = (w - 4)^2 + 3

# Plot C
plot(C, 
    xlims = (-8, 15),
    xaxis = "w",
    ylims = (-5, 100),
    yaxis = "Cost C(w)",
    legend = false)
```

---

For our test function we will use $C(w) = (w-4)^2 + 3$. It is clear that our *global minimizer* is attained at $w = 4$. But what if we didn't know this answer? Maybe we could take a guess? Suppose we initially guess $w_0 = 12.0$. See the following plot. 

---

```julia
# Make initial guess 
w_0 = 12.0

println("w_0 = $w_0")
println("")

scatter!([(w_0, C(w_0))], 
        color = "yellow")
```

---

Now that we have made a guess at our minimizer, we should decide to either move to the right or the left, or stay where we are. Recall from Calculus that we are at a local minimizer whenever the gradient (or derivative) is zero. For our simple test function we observe $C'(w) = 2w-8$. Further, $C'(w_0) = 6.0 > 0$, and hence we are not at a local minimizer. Choose $\alpha = .34$, and take a step in a descent direction. Our hope is to arrive at a new value $w_{k+1}$ for which $|\nabla C(w_{k+1})| < \epsilon$, where $\epsilon$ is some small positive number.  

---

```julia
# Define the derivative of C(w)
∇C(w) = 2.0*w - 8.0

# Set your step length
α = .34

# Apply gradient descent 
w_1 = w_0 - α*∇C(w_0)


println("w_0 = $w_0 and |∇C(w_0)| = ", abs(∇C(w_0)))
println("")
println("w_1 = $w_1 and |∇C(w_1)| = ", abs(∇C(w_1)))
println("")

plot!([(w_0, C(w_0)), (w_1, C(w_1))],
        color = "lightblue")
scatter!([(w_0, C(w_0)), (w_1, C(w_1))],
        color = "yellow")
```

---

After taking a step in a descent direction we have now found $w_1 = 6.56$. At this point $|\nabla C(w_1)| = 5.119$. Hence, we take another step in a descent direction. 

---

```julia
# Apply gradient descent 
w_2 = w_1 - α*∇C(w_1)

println("w_0 = $w_0 and |∇C(w_0)| = ", abs(∇C(w_0)))
println("")
println("w_1 = $w_1 and |∇C(w_1)| = ", abs(∇C(w_1)))
println("")
println("w_2 = $w_2 and |∇C(w_2)| = ", abs(∇C(w_2)))
println("")

plot!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2))],
        color = "lightblue")
scatter!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2))],
        color = "yellow")
```

---

After taking a step in a descent direction we have now found $w_2 = 4.819$. At this point $|\nabla C(w_2)| = 1.638$. Hence, we take another step in a descent direction.

---

```julia
# Apply gradient descent 
w_3 = w_2 - α*∇C(w_2)

println("w_0 = $w_0 and |∇C(w_0)| = ", abs(∇C(w_0)))
println("")
println("w_1 = $w_1 and |∇C(w_1)| = ", abs(∇C(w_1)))
println("")
println("w_2 = $w_2 and |∇C(w_2)| = ", abs(∇C(w_2)))
println("")
println("w_3 = $w_3 and |∇C(w_3)| = ", abs(∇C(w_3)))
println("")

plot!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2)), (w_3, C(w_3))],
        color = "lightblue")
scatter!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2)), (w_3, C(w_3))],
        color = "yellow")
```

---

After taking a step in a descent direction we have now found $w_3 = 4.262$. At this point $|\nabla C(w_3)| = 0.5252$. Just by evaluating the absolute value of the gradient, it seems that we are approaching a minimum. However, let's take one more step in a descent direction.

---

```julia
# Apply gradient descent 
w_4 = w_3 - α*∇C(w_3)

println("w_0 = $w_0 and |∇C(w_0)| = ", abs(∇C(w_0)))
println("")
println("w_1 = $w_1 and |∇C(w_1)| = ", abs(∇C(w_1)))
println("")
println("w_2 = $w_2 and |∇C(w_2)| = ", abs(∇C(w_2)))
println("")
println("w_3 = $w_3 and |∇C(w_3)| = ", abs(∇C(w_3)))
println("")
println("w_4 = $w_4 and |∇C(w_4)| = ", abs(∇C(w_4)))
println("")

plot!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2)), (w_3, C(w_3)), (w_4, C(w_4))],
        color = "lightblue")
scatter!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2)), (w_3, C(w_3)), (w_4, C(w_4))],
        color = "yellow")
```

---

At our fourth iteration we have found a sufficiently small output from calculating $|\nabla C(w_4)| = 0.167$. Numerically this means that we are close to a local extrema. This plot above varifies this assumption. So that's it! This idea of iteratively updating our input guess by way of stepping in a descent direction is the basic idea behind gradient descent. In the next notebook we will discuss problems that can occur when choosing your step length $\alpha_k$.
