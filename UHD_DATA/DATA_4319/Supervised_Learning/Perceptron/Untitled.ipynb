```julia
using Plots
theme(:dark)
```

```julia
function ∇f(x, ϵ)
    return (f(x + ϵ) - f(x))/ϵ
end
```

```julia
function gradient_descent_points(f, x_0, α, iteration_count)
    points = [(x_0, f(x_0))]
    for _ = 1:iteration_count
        x_0 = x_0 - α*∇f(x_0, 0.001)
        push!(points, (x_0, f(x_0)))
    end
    return points
end
```

```julia
f(x) = (x - 4)^2 + 3
iterations = 10
X = gradient_descent_points(f, 9.0, .75, iterations)
plot(f, 0, 10, 
    label = "f(x) = (x-4)^2 + 3", 
    title = "Gradient Descent Example")
for i = 1:iterations
    scatter!(X[i], label = "iteration $i")
end
scatter!()
```
